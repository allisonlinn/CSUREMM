```python
import pandas as pd
from google.colab import files

uploaded_files = files.upload()
file_name = next(iter(uploaded_files))  # Assuming you uploaded only one file
data = pd.read_csv(file_name)
##data = pd.read_excel("/content/drive/MyDrive/merged_data.xlsx")
##data.head()
```

```python
data = data.fillna('')

# Extract the 'Headline' column
headlines = data['Headline'].tolist()

# Perform the query on the 'Headline' column
output = query({
    "inputs": headlines
})

print(type(output))
print(output)

# Extract label and score from the output
labels = []
scores = []
for result in output:
    labels.append(result[0]['label'])
    scores.append(result[0]['score'])

# Create the new DataFrame
new_data = pd.DataFrame({'headline': headlines, 'date': data['Date'], 'label': labels, 'score': scores})

# Remove rows with 'no' in the label column
new_data = new_data[new_data['label'] != 'no']

# Filter rows where the score is less than 0.50
new_data = new_data[new_data['score'] >= 0.50]

# Save the DataFrame as a CSV file
new_data.to_csv('partialoutput.csv', index=False)

# Download the CSV file
files.download('partialoutput.csv')
```
